```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 25})
plt.style.use('ggplot')
import os
```

```python
ROOT = "../data/"
PATH = os.listdir("../data")
```

```python
PATH = [p for p in PATH if pd.read_csv(ROOT+p,index_col=0).shape[0]>0]
```

```python
save_path = '../graphs/per_country/'
```

```python
for p in PATH[:3]:
    fig, axes = plt.subplots(figsize=(12,8))
    fig.subplots_adjust(hspace=0.3)
    fig.subplots_adjust(top=0.7)
    data = pd.read_csv(ROOT+p, index_col = 0, names = ['confirmed','dead','recovered','active'],skiprows=1)
    inx = np.array([ "/".join(i.split("-")[:2]) for i in data.index])
    data.set_index(inx,inplace=True)
    data = data.iloc[::-1]
    data.plot(ax = axes, title = p.split(".")[0].split("_")[0],marker='o') 
    axes.set_title(p.split(".")[0],fontweight="bold", size=18) # Title
    axes.set_ylabel('Cases', fontsize = 20) 
    axes.set_xlabel('Date', fontsize = 20) # X label
    fig.tight_layout()
    plt.savefig(save_path+p.split(".")[0]+'.png',quality = 95,orientation='landscape',dpi=72)
```
